Precisamos imprimir um número para cada andar de um hotel de 20 andares. Porém, o dono do hotel é supersticioso e optou por não ter um 13º andar.

1 - Escreva um código que imprima todos os números exceto o número 13.

2 - Escreva mais um código que resolva o mesmo problema, mas dessa vez usando o laço de repetição 'while'.

3 - Como desafio, imprima eles em ordem decrescente (20, 19, 18...)

#1 - Laço de repetição for

Comment: Como o térreo não foi mencionado, mas apenas os andares, então iniciei o laço na posição 1 com o 1º andar, repetindo o print até o 21º andar, já que a numeração do 13º andar foi pulada pois o dono do hotel é supersticioso, mas o prédio ainda continuou com 20 andares. E para isso coloquei o limite da contagem como 22, pois ele sempre para o print na posição anterior, neste caso, na 21ª.

```python
for i in range(1,22):
  if(i == 13):
    continue               #comment -> Essa key-word continue: interrompe a laço de repetição e pula para o próximo item.
  else:
    print(i)
```

#2 - Laço de repetição while

Comment: Como o térreo não foi mencionado, mas apenas os andares, então iniciei o laço na posição 1 com o 1º andar, repetindo o print até o 21º andar, já que a numeração do 13º andar foi pulada pois o dono do hotel é supersticioso, mas o prédio ainda continuou com 20 andares. E para isso coloquei o limite da contagem como 22, pois ele sempre para o print na posição anterior, neste caso, na 21ª.

```python
contador = 1

while (contador <=21):
  if(contador == 13):
    contador = contador + 1
    continue
  else:
    print(contador)
    contador = contador + 1
```

#3 - Desafio com laço for

Comment: Como o desafio pede o print pela ordem decrescente, dessa vez já iniciei pelo último andar, o 21, já que o dono do hotel é supersticioso e vai querer pular a numeração do 13º andar, caindo de 1 em 1, até o limite 0, pois a repetição sempre vai parar na posição interior ao limite, que neste caso será a do 1º andar. Térreo não foi mencionado, apenas os andares, então ignorei.



```python
print("Com laço de repetição for:")
for i in range(21,0,-1):
  if(i == 13):
    continue               #comment -> Essa key-word continue: interrompe a laço de repetição e pula para o próximo item.
  else:
    print(i)

print("Com laço de repetição while")
contador = 21

while (contador >0):
  if(contador == 13):
    contador = contador - 1
    continue
  else:
    print(contador)
    contador = contador -1
```
